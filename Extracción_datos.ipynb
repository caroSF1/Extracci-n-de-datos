**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Se realiza un filtro por objeto, donde se toma la columna de NOMBRE_CLIENTE,  y nos mostrará los datos que sean iguales a PÚBLICO EN GENERAL
filtro1=df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro1
```

```python
#Filtro por cabecera, se filtraron las primeras 14 columnas
filtro5=df.head(15)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparación, se comparó que el precio unitario fuera menor que 400
filtro6=df[df["PRECIO_UNITARIO"] < 400]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y, se filtro con las condiciones a cumplir que el precio unitario fuera menor a 500 y que la fecha de documento sea de 2022-12-08
filtro7=df[(df["PRECIO_UNITARIO"] < 500) & (df["FECHA_DOC"] == "2022-12-08")]
filtro7
```

```python
#Filtro O, se filtro con las condiciones a cumplir que el precio unitario fuera menor a 700 y que la fecha de documento sea de 2022-12-08, 
#en este caso, puede ser uno o otro, no debe cumplir con las dos condiciones
filtro8=df[(df["PRECIO_UNITARIO"] < 700) | (df["FECHA_DOC"] == "2022-12-08")]
filtro8
```

```python
#Filtro not, te muestra los datos que sean diferentes a un precio unitario mayor o igual a 500 y una fecha de documento diferente a 2022-12-08
filtro9=df[~(df["PRECIO_UNITARIO"] >= 500) & ~(df["FECHA_DOC"] == "2022-12-08")]
filtro9
```

```python
#Se realiza un filtro por objeto, donde se toma la columna de NOMBRE_CLIENTE,  y nos mostrará los datos que sean iguales a PÚBLICO EN GENERAL
filtro1.to_csv("Filtro por objeto.csv")
```
