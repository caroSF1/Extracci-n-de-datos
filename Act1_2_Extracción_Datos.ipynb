**1. CVE_CLPV(VALORES 1000:2000)**

```python
fil1=df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
fil1
```

**2.CVE_VEND(TODAS LAS CLAVES EXCEPTO "5" Y "4")**

```python
fil2=df[~(df["CVE_VEND"] == "5") & ~(df["CVE_VEND"] == "4")]
fil2
```

**3. FECHA_ENT   (Fechas de ‚Äú28/02/2022‚Äù)**

```python
fil3=df[df["FECHA_ENT"] == "28/02/2022  0:00:00"]
fil3
```

**4. CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (‚ÄúE‚Äù)**

```python
fil4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
fil4
```

**5. Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT**

```python
fil5=df.iloc[:,[0, 6, 7, 9]] 
fil5
```

**6. Solo las filas de 7001-7099**

```python
fil6 = df.iloc[ [7001, 7099] , : ]
fil6
```

**7. Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)**

```python
df1= pd.read_excel('Datos de FacturacioÃÅn.xlsx', index_col=3)
df1
```

```python
fil7=df1.loc[[1.0,2.0], ["FECHAELAB"]]
fil7
```
